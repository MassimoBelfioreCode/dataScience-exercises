 - <b>DataFrame</b>: Una struttura <b>dati tabellare</b> bidimensionale <b>simile a un foglio di calcolo</b>, composta da righe e colonne. È usato per rappresentare e manipolare dati in formato tabellare.

### Lettura file excel

```python
import pandas as pd

coding     = pd.read_excel("File/CSV/liste coding e non-coding RNA.xlsx", sheet_name="posizioni coding RNA")
non_coding = pd.read_excel("File/CSV/liste coding e non-coding RNA.xlsx", sheet_name="posizioni non-coding RNA")
```

#### Eliminazione colonna name in entrambi i DataFrame

```python
coding.drop(columns="name", inplace=True)
non_coding.drop(columns="name", inplace=True)
```

#### Rinominare name2 in Coding (dataframe coding) e NonCoding (dataframe non_coding)

```python
coding.rename(columns={"name2": "coding"}, inplace=True)
non_coding.rename(columns={"name2": "non_coding"}, inplace=True)
```

#### primi 10000 campioni

```python
coding = coding[:10000]
non_coding = non_coding[:10000]
```

#### Join tra due DataFrame
merge(right, how='inner', on=None, left_on=None, right_on=None, ...)
 - how: {‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}, default ‘inner’
 - on: nome della colonna (in entrambi i df) che deve essere utilizzata per la join.
 - left_on, right_on: sono usati in sostituzione a on se le colonne nei due df hanno nomi differenti.

```python
coding_no_coding = pd.merge(coding, non_coding, on="chrom")
```

```python
coding_no_coding
```

#### Comprensione del filtraggio
 - garantiamo che stand_x e stand_y siano di segno opposto

```python
condizione = coding_no_coding["strand_x"] != coding_no_coding["strand_y"]
coding_no_coding[condizione]
```

#### Utilizzo funzione apply

<b>NOTA</b>:
     
     + indica txStart_x ----------> txEnd_x
     - indica txStart_y <----------	txEnd_y

Vogliamo tenere tutte le righe della tabella tale che + e - siano totalmente o parzialmente sovrapposti

<b>Esempio</b>:
     
     + indica  ----------> 
     - indica       <----------	


```python
def sovrapposizione(row):
    if row["txStart_y"] < row["txEnd_x"] <= row["txEnd_y"]:
        return row["txEnd_x"] - row["txStart_y"]
    if row["txStart_x"] < row["txEnd_y"] <= row["txEnd_x"]:
        return row["txEnd_y"] - row["txStart_x"]
    return -1

coding_no_coding["sovrapposizione"] = coding_no_coding.apply(lambda x: sovrapposizione(x), axis=1)
```

```python
# loc -> selezionare righe e colonne in un DataFrame utilizzando etichette o etichette booleane.
coding_no_coding = coding_no_coding.loc[coding_no_coding["sovrapposizione"] >= 0]
```

```python
coding_no_coding
```

#### Selezioniamo un sott'insieme di colonne (teniamo solo chrom, coding, non_coding, sovrapposizione)

```python
coding_no_coding = coding_no_coding[["chrom", "coding", "non_coding", "sovrapposizione"]].reset_index(drop=True)
```

```python
coding_no_coding
```

#### Group by

```python
coding_no_coding.groupby(
    by=["chrom", "coding", "non_coding"]
).agg({'sovrapposizione': ['mean', 'min', 'max']})
```
